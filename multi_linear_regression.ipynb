```python
reg=linear_model.LinearRegression()
```

```python
reg.fit(df[['age','height','weight']],df['premium'])
```

```python
reg.coef_
```

```python
reg.intercept_
```

```python
# Whose age is 27 height as 165.56 and weight is 60,
reg.predict([[27,167.56,60]])
```

```python
 #from reg formula
5.*27+0.*167.56+5.*60
```

```python
#Whose age is 60, height is 165.10 and weight is 80.
reg.predict([[60,165.10,80]])
```

```python
 #from reg formula
5.*60+0.*165.10+5.*80
```
